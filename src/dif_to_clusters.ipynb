```python
import pandas as pd

data = pd.read_csv('../../data/datathon_data_one_hot_encoded.csv')
anomal_data = data[data['label_regular'] == 0]
anomal_data
```

```python
anomal_data_no_na = anomal_data.loc[:, (anomal_data != 0).any(axis=0)]
anomal_data_no_na
```

```python
categorical_values = anomal_data_no_na.drop(columns=['DMBTR', 'WRBTR'])
categorical_values
```

```python
from sklearn.preprocessing import MinMaxScaler

# Assuming anomal_data_no_na is already defined
anomal_data_dmbtr = anomal_data_no_na[['DMBTR']]

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the data
anomal_data_dmbtr_normalized = scaler.fit_transform(anomal_data_dmbtr)

anomal_data_no_na['DMBTR'] = anomal_data_dmbtr_normalized

anomal_data_dmbtr_normalized
```

```python
anomal_data_no_na
```

```python
import pandas as pd
from sklearn.cluster import DBSCAN

anomal_data_dmbtr_normalized = pd.DataFrame(anomal_data_dmbtr_normalized)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(anomal_data_dmbtr_normalized)
anomal_data_dmbtr_normalized['cluster'] = clusters

# Count the number of clusters (excluding noise points)
num_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)

# Display the number of clusters
print(f'Number of clusters: {num_clusters}')
```

```python
anomal_data_dmbtr
```

```python
# Calculate centroids for each cluster
centroids = anomal_data_dmbtr_normalized.groupby('cluster').mean()
centroids
```

```python
import matplotlib.pyplot as plt

# Plot the centroids
plt.figure(figsize=(10, 6))
for i, centroid in centroids.iterrows():
    plt.plot(centroid, marker='o', label=f'Cluster {i}')

plt.xlabel('Features')
plt.ylabel('Centroid Values')
plt.title('Centroids of Each Cluster')
plt.legend()
plt.show()
```

```python
categorical_averages = pd.DataFrame(categorical_values.mean())
categorical_averages.T
```

```python
centroids_data = {
    'dmtbr1': centroids.iloc[0],
    'dmtbr2': centroids.iloc[1]
}

# Create DataFrame
centroids_data_df = pd.DataFrame(centroids_data).T
centroids_data_df
```

```python
categorical_averages.columns = ['value']
centroids_data_df.columns = ['value']

benchmark_data = pd.concat([categorical_averages, centroids_data_df])
benchmark_data
```

```python
target_correlation = pd.read_csv('../../data/target_correlation.csv')
target_correlation.drop(target_correlation.index[1])
```

```python
benchmark_data.to_csv('../../data/benchmark_data.csv', index=False)
```

```python
anomal_data_entry = pd.DataFrame(anomal_data_no_na.loc[12939])
anomal_data_entry.drop(anomal_data_entry.index[1])
anomal_data_entry
```

```python
differences = {} # Will store the differences as a rank: difference dictionary

for row in target_correlation.iterrows():
    try:
        key = row[1]['type_value']
        if 'dmtbr1' in key:
            differences['dmtbr1'] = abs(anomal_data_entry.loc[key].iloc[0] - benchmark_data.loc[key][0])
        elif 'dmtbr2' in key:
            differences['dmtbr2'] = abs(anomal_data_entry.loc[key].iloc[0] - benchmark_data.loc[key][0])
        else:
            differences[key] = abs(anomal_data_entry.loc[key].iloc[0] - benchmark_data.loc[key][0])
    except KeyError as ke:
        print(f'KeyError: {ke}')
        continue
```

```python
pd.DataFrame(differences, index=[0]).T
```

```python
# Sort the differences dictionary by its values
sorted_differences = dict(sorted(differences.items(), key=lambda item: item))

pd.DataFrame(sorted_differences, index=[0]).T
```

```python
import pandas as pd

data = pd.read_csv('../data/datathon_data.csv')
```

```python
categorical_features = data.select_dtypes(include=['object']).columns
categorical_features
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, drop='first')
encodings = encoder.fit_transform(data[categorical_features])
encodings
```

```python
encoded_columns = encoder.get_feature_names_out(categorical_features)

encoded_df = pd.DataFrame(encodings, columns=encoded_columns)

data_dropped = data.drop(categorical_features, axis=1)
data_dropped = data_dropped.drop('BELNR', axis=1)

data_one_hot_encoded = pd.concat([data_dropped, encoded_df], axis=1)
data_one_hot_encoded
```

```python
data_one_hot_encoded.to_csv('../data/datathon_data_one_hot_encoded.csv', index=False)
```

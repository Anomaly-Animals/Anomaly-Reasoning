```python
# Calculate the ratio first
data['DMBTR_WRBTR_ratio'] = data['WRBTR'] / data['DMBTR'].replace(0, 1e-10)

# Create the filter condition
ratio_filter = (data['DMBTR_WRBTR_ratio'] >= 0.059) & (data['DMBTR_WRBTR_ratio'] <= 0.082) & (data['DMBTR'] > 900000) & (data['DMBTR'] < 1000000)

# Display the filtered data
golden_ratio_data = data[ratio_filter]

golden_ratio_data
```

```python
# Convert to string and check decimal part length
def has_more_than_2_decimals(value):
    if pd.isna(value):
        return False
        
    # Convert to string representation
    str_value = str(float(value))
    
    # Check if there's a decimal point and more than 2 digits after it
    if '.' in str_value:
        decimals = str_value.split('.')[1]
        # Remove trailing zeros
        decimals = decimals.rstrip('0')
        return len(decimals) > 2
    return False

# Apply the function to filter the data
filtered_data = data[data['DMBTR'].apply(has_more_than_2_decimals)]

# Display results
print(f"Found {len(filtered_data)} rows where DMBTR has more than 2 decimal places")

filtered_data
```
